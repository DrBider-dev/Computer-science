# Euclidian GCD

This is an example of the __Euclidean Algorithm__ for this course

```python
def euclidean_test(a: int, b: int) -> int:
    """
    This is a function to implement the cassical euclidean algorithm
    to calculate the GCD.

    Args:
        a(int): First number, should be greater than b
        b(int): Aecond number

    Returns:
        An integer with the GCD for the parameters provided.
    """

    if a <= b:
        a, b = b, a

    while b != 0:
        r = a % b
        a, b = b, r
    return a


result = euclidean_test(90, 30)

print(result)
```

# Max in an Array


```python
def maxArray(array: list) -> int:
    n = len(array)
    if n == 0:
        raise ValueError("Array must have elements")
    max = array[0]
    for i in range(1, n):
        if max < array[i]:
            max = array[i]
    
    return max

array = (1,2,3,34,5)

a = maxArray(array)
print(a)
```

# Recursive Euclidean GCD

```python
def euclidean_test_recursive(a: int, b: int) -> int:
    """
    This is a function to implement the cassical euclidean algorithm
    to calculate the GCD.

    Args:
        a(int): First number, should be greater than b
        b(int): Aecond number

    Returns:
        An integer with the GCD for the parameters provided.
    """

    if b == 0:
        return a
    else:
        return euclidean_test_recursive(b, a % b)


result = euclidean_test_recursive(90, 30)

print(result)
```

# Decision Tree Example

```python
def cast_num_to_letter_grade(score: int) -> str:
    if 0 <= score <= 100:
        grade = None
        if score >= 90:
            grade = 'A'
        elif score >= 80:
            grade = 'B'
        elif score >= 70:
            grade = 'C'
        elif score >= 60:
            grade = 'D'
        else:
            grade = 'F'
        return grade
    else:
        raise ValueError("Error")
    
print(cast_num_to_letter_grade(100))
```
