# Euclidean GCD Algorithm

This is an algorithm used to find the __Greatest Common Divisor__ between 2 numbers

```python
def euclidean_test(a: int, b: int) -> int:
    """
    This is a function to implement the classical euclidean algorithm
    to calculate the GCD.

    Args:
        a(int): First number, should be greater than b
        b(int): Second number

    Returns:
        An integer with the GCD for the parameters provided.
    """

    if (a <= b):
        a, b = b, a

    while (b != 0):
        r = a % b
        a, b = b, r
    return a


result = euclidean_test(90, 30)
print(result)
```

# Index sum

This is an algorithm that returns de indexes of two number that add up to a given number

```python
def indexes(array: list, number: int) -> list:
    """
    This is a function that recieves an array and a number, 
    and returns the indexes of two numbers in the array
    that add up to the target number.

    Args:
        array(list): The list of integers to search through.
        number(int): The target sum to find in the array.
    """
    diccionary = {}
    if len(array) == 0:
        raise ValueError("There are no elements in the list")

    for i in range(len(array)):
        key = number - array[i]
        
        if key in diccionary:
            return [diccionary[key], i]
        
        diccionary[array[i]] = i
    raise ValueError("There are no two numbers that add up to the target")

print(indexes([1, 2, 3], 3))
```

# Square root calculation

The main idea of this algorithm is to aproximate the square root of a positive number by using __the babylonian method__.

```python
def square_root(number: float) -> float:
    """
    This is a function that calculates the square root of a number

    Args:
        number(int): The number to calculate the square root of.

    Returns:
        An integer with the approximate square root of the number.
    """

    if number < 0:
        raise ValueError("No negatives allowed")
    if number == 0 or number == 1:
        return number
    
    diff= 0.01
    x = number/2
    while True:
        x_next = 0.5 * (x + number / x)
        diff_temp = abs(x_next - x)
        x = x_next
        if diff_temp < diff:
            break
    return x

print(square_root(144))
```

# Maximum number in an array

This algorithm finds the greatest number in an array of integers.

```python
def maximum(array: list) -> int:
    """
    This is a fuction that calculates the maximum number in an array.

    Args:
        array(list): The list of integers to search through.
    Returns:
        An integer with the maximum number in the array.
    """
    if len(array) == 0:
        raise ValueError("The array is empty")

    max_number = array[0]

    for i in range(len(array)):
        if max_number < array[i]:
            max_number = array[i]
    return max_number

print(maximum([1,5,7,14,3,26,8,25,26,0]))
```

# Decision Tree Example

```python
def cast_num_to_letter_grade(score: int) -> str:
    if 0 <= score <= 100:
        grade = None
        if score >= 90:
            grade = 'A'
        elif score >= 80:
            grade = 'B'
        elif score >= 70:
            grade = 'C'
        elif score >= 60:
            grade = 'D'
        else:
            grade = 'F'
        return grade
    else:
        raise ValueError("Error")
    
print(cast_num_to_letter_grade(100))
```

```python
def maxArray(array: list) -> int:
    n = len(array)
    if n == 0:
        raise ValueError("Array must have elements")
    max = array[0]
    for i in range(1, n):
        if max < array[i]:
            max = array[i]
    
    return max

array = (1,2,3,34,5)

a = maxArray(array)
print(a)
```

# Recursive Euclidean GCD

```python
def euclidean_test_recursive(a: int, b: int) -> int:
    """
    This is a function to implement the cassical euclidean algorithm
    to calculate the GCD.

    Args:
        a(int): First number, should be greater than b
        b(int): Aecond number

    Returns:
        An integer with the GCD for the parameters provided.
    """

    if b == 0:
        return a
    else:
        return euclidean_test_recursive(b, a % b)


result = euclidean_test_recursive(90, 30)

print(result)
```
