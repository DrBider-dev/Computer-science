# __Search Algorithms__

- linear serach: dahh (is not divide and conquer)



```python
from typing import Any
def linear_search(a: list, x: Any) -> int:
    """
    This function represents the behavior of a linear search
    
    Args:
        a(list): Set of elements
        x(Any): Element to be searched

    Returns:
        An integer number with the index where x appears. -1 if not found
    """

    index = -1
    for i, item in enumerate(a):
        if item == x:
            index = i
            break
    return index
```

# Find Person's phone

```python
directory = [
    {
        'name': "Brayan",
        'number': 12345
    },
    {
        'name': "Samuel",
        'number': 23423
    },
    {
        'name': "Jan",
        'number': 89765
    }

]
```

```python
def find_phone_number(directory: dict, name: str) -> int:
    """
    This function finds a Person's phone by their name

    Args:
        directory(dict): directory with the names and phones
        name(str): The person's name we're looking for

    Returns:
        The phone of the name
    """
    result = None
    for person in directory:
        if person['name'] == name:
            result = person['number']
            break
    return result
```

```python
name = 'Brayan'
print(find_phone_number(directory, name))
```

# Binary Search
Hell yeah

```python
def binary_search(a: list, x: int) -> int:
    """
    This function finds an element in a given list

    Args:
        a(list): the list of elements
        x(int): the element we're looking for

    Returns:
        The index of the element in the list. -1 if not found
    """

    a = sorted(a)
    low = 0
    high = len(a) - 1
    index = -1

    while low <= high:
        mid = (low  + high)//2
        if a[mid] == x:
            index = mid
            break
        elif a[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return index
```

```python
a = [i for i in range(1,101)]
x = 41
print(linear_search(a,x))
print(binary_search(a,x))
```
