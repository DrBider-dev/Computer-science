# Brute force: Exhaustive Search

it tries all possible solutions

- Exhaustive: Examines every possiblity
- Expensive: Often exponential time complexity and resources
- Always finds optimal solutions
- Ez to understand

### When to use it:
- small problems
- No efficient algorithm is known
- When optimization overhead exceeds brute force cost

## Password Cracking

```python
def get_possibilities(character_set: list, length: int, current: str = ""):
    if length == 1:
        return [current + c for c in character_set]
    else:
        result = []
        for c in character_set:
            result.extend( get_possibilities(character_set, length - 1, current + c))
```

```python
def get_password(character_set: list, password: str, max_length: int) -> str:
    result = None
    for length in range(1, max_length + 1):
        possibilities = get_possibilities(character_set, length)
        for candidate in possibilities:
            if candidate == password:
                print("Found")
                result = candidate
    return result

    character_set = list('abcdefghijklmnopqrstuvwxyz')
    max_ = 8
    password = 'zefhusi'
    get_password(character_set, max_, password)
```
