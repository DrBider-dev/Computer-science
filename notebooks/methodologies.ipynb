# Greedy Method

loof for easy solutions

```python
def coin(amount: int, currency: list) -> list:
    currency = sorted(currency, reverse=True)
    result = []
    for coin in currency:
        if amount >= coin:
            coin_num = amount // coin
            amount = amount % coin
            result.append((coin, coin_num))
    return result

print(coin(25950, [50, 200, 100, 1000, 500]))
```

### __Actually this is a greedy__

```python
def knapsack(capacity, items):
    items = sorted(items, key=lambda item:item[1] / item[0], reverse=True)
    result = []
    for item in items:
        if item[0] <= capacity:
            result.append(item)
            capacity -= item[0]
    print(f'La maleta tiene {len(result)} items con un valor de\
          {sum( [item[1] for item in result])}')
    return result

knapsack(7, items)
```

# Queen problem

```python
data = input()
while data != '0 0 0 0':
    x1, y1, x2, y2 = map(int, data.split())
    result = 2
    if x1 == x2 and y1 == y2:
        result = 0
    elif (x1 == x2) or (y1 == y2) or (abs(x1 - x2) == abs(y1 - y2)):
        result = 1
    print(result)
    data = input()
```

# Optical reader

```python
n = int(input())
grades = ['A', 'B', 'C', 'D', 'E']
while n != 0:
    for sheet in range(n):
        answers = list(map(int, input().split()))
        response = '*'
        found = False
        for i, answer in enumerate(answers):
            if answer <= 127:
                if not found:
                    found = True
                    response = grades[i]
                else:
                    response = '*' #two answers
                    break
        print(response)

    n = int(input())
```
