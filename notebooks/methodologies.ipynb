# Greedy Method

loof for easy solutions

```python
def coin(amount: int, currency: list) -> list:
    currency = sorted(currency, reverse=True)
    result = []
    for coin in currency:
        if amount >= coin:
            coin_num = amount // coin
            amount = amount % coin
            result.append((coin, coin_num))
    return result

print(coin(25950, [50, 200, 100, 1000, 500]))
```

# Divide and Conquer

Split the problem into smaller 'sub-problems' solve them and combine the solutions

## knapsack problem (0/1)


```python
# items -> weight, value
items = [(5,6, 'Items 1'),
            (7,8, 'Items 2'),
            (8,9, 'Items 3'),
            (9,10, 'Items 4'),
            (1,2, 'Items 5')]

knapsack = 120  # Capacity

```

```python
def knapsack(capacity, items):
    items = sorted(items, key=lambda item:item[1] / item[0], reverse=True)
    result = []
    for item in items:
        if item[0] <= capacity:
            result.append(item)
            capacity -= item[0]
    print(f'La maleta tiene {len(result)} items con un valor de\
          {sum( [item[1] for item in result])}')
    return result

knapsack(1800000, items)

# mirar que esta mal
```
